```python
import numpy as np
f = lambda x1, x2: (x1 - 3)**2 + (x2 - 2)**2
normalize = lambda x: x / np.linalg.norm(x)
gf = lambda x1, x2: normalize(np.array([2*(x1-3), 2*(x2-2)])) #gradient of f
g1 = lambda x1, x2: x1**2 + x2**2 - 5 #constraint1 g1(x1,x2)<=0
gg1 = lambda x1, x2: normalize(np.array([2*x1,2*x2]))
g2 = lambda x1, x2: x1 + 2*x2 -4
gg2 = lambda x1, x3: normalize(np.array([1,2]))
g3 = lambda x1, x2: -x1
g4 = lambda x1,x2: -x2
x0 = np.array([2,1])
xx = np.array([3,2])
```

```python
#setting up the plotting area
import matplotlib.pyplot as plt
fig,ax = plt.subplots()
ax.set_aspect('equal')
ax.set_xlim(0,3.5)
ax.set_ylim(0,2.5)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.set_xlabel(r'$x_1$')
ax.set_ylabel(r'$x_2$')

#prepare data
x1 = np.linspace(0,3.5,500)
x2 = np.linspace(0,2.5,500)
X1,X2 = np.meshgrid(x1,x2)
F = f(X1,X2)
G1 = g1(X1,X2)
G2 = g2(X1,X2)
G3 = g3(X1,X2)
G4 = g4(X1,X2)

#plot the contour of the objective function
ax.contour(X1,X2, F, levels=[0.1,0.5,1,2], cmap='Greys')
#plot the point in discussion
ax.plot(x0[0],x0[1],'ro',markersize=5)
ax.plot(xx[0],xx[1], 'mo', markersize=5)
#plot the feasible region
plt.imshow(((G1<=0) & (G2<=0) & (G3<=0) & (G4<=0)).astype(int),
           extent=(X1.min(),X1.max(),X2.min(),X2.max()), origin = 'lower', cmap='Greys', alpha=0.4)

#plot gradients
plt.annotate(r"$\nabla f$",xy=x0, xytext=x0 + gf(*x0),arrowprops={'arrowstyle':'<-','color':'r','lw':1},va='center',ha='center')
plt.annotate(r"$-\nabla g_1$",xy=x0, xytext=x0 - gg1(*x0),arrowprops={'arrowstyle':'<-','color':'g','lw':1},va='center',ha='center')
plt.annotate(r"$- \nabla g_2$",xy=x0, xytext=x0 - gg2(*x0),arrowprops={'arrowstyle':'<-','color':'g','lw':1},va='center',ha='center')
```
