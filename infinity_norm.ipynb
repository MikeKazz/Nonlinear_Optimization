**Example 1.** Consider the following overdetermined system of linear equations Ax=b

```python
import numpy as np

A = np.array([[1,1],
             [1,-1],
             [1,2],
             [2,1],
             [3,1]])

b = np.array([3,1,7,8,6])

# we first solve it with Euclidean norm
x, residuals, rank, s = np.linalg.lstsq(A,b, rcond= None)
print(x)
```

# **Visualization of the problem**

```python
import matplotlib.pyplot as plt

X = np.linspace(0,5.0,4)
y1 = 3-X
y2 = X-1
y3 = 0.5*(7-X)
y4 = 8-2*X
y5 = 6-3*X

plt.plot(X,y1)
plt.plot(X,y2)
plt.plot(X,y3)
plt.plot(X,y4)
plt.plot(X,y5)

plt.plot(x[0],x[1],'ro')
plt.annotate(r'solution',xy=(x[0],x[1]),xycoords='data',xytext=(-70,-25), textcoords='offset points',fontsize=12,arrowprops=dict(arrowstyle='->',connectionstyle='arc3,rad=.2'))
```

# **2-norm**

```python
#for 2-norm the objective function is: 0.5 * [(3-x-y)**2 + (1-x+y)**2 + (7-x-2*y)**2 + (8-2*x-y)**2 + (6-3*x-y)**2] 
f2 = lambda x,y: 0.5 * ((3-x-y)**2 + (1-x+y)**2 + (7-x-2*y)**2 + (8-2*x-y)**2 + (6-3*x-y)**2) 


from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = Axes3D(fig)

X = np.arange(0,4,0.1)
Y = np.arange(-6,8,0.1)
X, Y =np.meshgrid(X,Y)
Z = f2(X,Y)

ax.plot_surface(X,Y,Z, rstride= 1, cstride= 1, cmap= 'Blues')
ax.contourf(X,Y,Z,zdir='z', offset=-2, cmap='Blues')
plt.show()
```

```python
#2-D contour

Z=f2(X,Y)
plt.contourf(X,Y,Z,8, alpha= 0.75, cmap='Blues')
C = plt.contour(X,Y,Z,10, colors='black', linewidth= 0.5)
plt.clabel(C, inline=1, fontsize= 10)
plt.show()

print('\nwe increse the resolution\n')

C = plt.contour(X,Y,Z,300, colors='black', linewidth= 0.5)
plt.clabel(C, inline=1, fontsize= 10)
plt.show()
```

## **1-norm**

```python
#for 1-norm the objective function is:
f1 = lambda x,y: (abs(3-x-y) + abs(1-x+y) + abs(7-x-2*y) + abs(8-2*x-y) + abs(6-3*x-y)) 

Z = f1(X,Y)
plt.contourf(X,Y,Z,8, alpha= 0.75, cmap='Blues')
C = plt.contour(X,Y,Z,15, colors='black', linewidth= 0.5)
plt.clabel(C, inline=1, fontsize= 10)
plt.show()

fig = plt.figure()
ax = Axes3D(fig)

X = np.arange(0,4,0.1)
Y = np.arange(-6,8,0.1)
X, Y =np.meshgrid(X,Y)


ax.plot_surface(X,Y,Z, rstride= 1, cstride= 1, cmap= 'Blues')
ax.contourf(X,Y,Z,zdir='z', offset=-2, cmap='Blues')
plt.show()
```

# **infinity-norm**

```python
#for infinity-norm the objective function is:
fmax = lambda x,y: np.amax(np.array([abs(3-x-y) ,abs(1-x+y) ,abs(7-x-2*y) ,abs(8-2*x-y),abs(6-3*x-y)]),axis=0) 

Z = fmax(X,Y)

fig = plt.figure()
ax = Axes3D(fig)
X = np.arange(0,4,0.1)
Y = np.arange(-6,8,0.1)
X, Y =np.meshgrid(X,Y)

ax.plot_surface(X,Y,Z, rstride= 1, cstride= 1, cmap= 'Blues')
ax.contourf(X,Y,Z,zdir='z', offset=-2, cmap='Blues')
plt.show()

plt.contourf(X,Y,Z,8, alpha= 0.75, cmap='Blues')
C = plt.contour(X,Y,Z,15, colors='black', linewidth= 0.5)
plt.clabel(C, inline=1, fontsize= 10)
plt.show()
```
